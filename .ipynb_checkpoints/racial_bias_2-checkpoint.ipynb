```python
classes = ['White', 'Black', 'Asian', 'Indian', 'Others']
```

```python
from tensorflow.keras.models import load_model

modelo_utkface_vgg16 = load_model('models/race_bias_classification_vgg16_utkface.keras')
```

```python
cm = confusion_matrix(y_true=test_batches.classes, y_pred=np.argmax(predictions, axis=-1))
```

```python
test_batches.class_indices
```

```python
def plot_confusion_matrix(cm, classes, normalize=True, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_mark = np.arange(len(classes))
    plt.xticks(tick_mark, classes, rotation=45)
    plt.yticks(tick_mark, classes)

    if normalize:
        cm = (cm.astype('float') / cm.sum(axis=1)[:, np.newaxis])*100
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, f'{cm[i, j]:.2f}%', horizontalalignment="center", color="white" if cm[i,j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.savefig(f'confusion_matrix/Vgg16/confusion_matrix_percent_UTKFace_vgg16_test_final.png', dpi=500, bbox_inches='tight')
```

```python
cm_plot_labels = ['White', 'Black', 'Asian', 'Indian', 'Others']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels)
```

```python
data_save = np.savetxt("Dados/Vgg16/UTKFace.dat", cm)
```

```python
train_path = 'UTKFace/'

classes = ['White', 'Black', 'Asian', 'Indian', 'Others']
```

```python
train_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.resnet50.preprocess_input) \
    .flow_from_directory(directory=train_path, target_size=(224,224), classes=classes, batch_size=10)
```

```python
resnet_model = tf.keras.applications.resnet50.ResNet50()

resnet_model.summary()

x = resnet_model.layers[-2].output
output = Dense(units=5, activation='softmax')(x)

model = Model(inputs=resnet_model.input, outputs=output)
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
epochs = 15
history = model.fit(x=train_batches, epochs=epochs, verbose=2)
```

### Saving Model

```python
if os.path.isfile('models/race_bias_classification_resnet_utkface_2.keras') is False:
    model.save('models/race_bias_classification_resnet_utkface_2.keras')
```

### Test Step

```python
classes = ['White', 'Black', 'Asian', 'Indian', 'Others']
test_path = 'Dataset_Fair_Face_test/'
test_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.resnet50.preprocess_input) \
    .flow_from_directory(directory=test_path, target_size=(224,224), classes=classes, batch_size=10, shuffle=False)
```

```python
from tensorflow.keras.models import load_model

modelo_utkface_resnet = load_model('models/race_bias_classification_resnet_utkface.keras')
```

```python
predictions = model.predict(x=test_batches, verbose=0)
```

```python
cm = confusion_matrix(y_true=test_batches.classes, y_pred=np.argmax(predictions, axis=-1))
```

```python
def plot_confusion_matrix(cm, classes, normalize=True, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_mark = np.arange(len(classes))
    plt.xticks(tick_mark, classes, rotation=45)
    plt.yticks(tick_mark, classes)

    if normalize:
        cm = (cm.astype('float') / cm.sum(axis=1)[:, np.newaxis])*100
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, f'{cm[i, j]:.2f}%', horizontalalignment="center", color="white" if cm[i,j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.savefig(f'confusion_matrix/ResNet50/confusion_matrix_percent_utkface_resnet_test_final_2.png', dpi=500, bbox_inches='tight')
```

```python
cm_plot_labels = ['White', 'Black', 'Asian', 'Indian', 'Others']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels)
```
